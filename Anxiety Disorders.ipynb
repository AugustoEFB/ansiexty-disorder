```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings("ignore")
 
```

```python
df = pd.read_excel('/Users/Augusto/Desktop/Test/Training_Data.xlsx')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
X = df.drop(['SAD', 'GAD',"Subject","Onset Child tries unsuccessfully to leave daycare/school due to anxiety","Onset Has to be taken to daycare/school because of separation anxiety","Onset Fear about calamitous separation","Onset Child leaves daycare/school due to anxiety"], axis=1)
y_SAD = df['SAD']
y_GAD = df['GAD']
```

```python
numerical_cols= X.select_dtypes(include=['int64','float64']).columns
```

```python
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
    ])
```

```python
model = RandomForestClassifier(n_estimators=917, random_state=0)
```

```python
pipeline_SAD = Pipeline(steps=[('preprocessor',preprocessor), ('model', model)])
```

```python
X_train_SAD, X_test_SAD, y_train_SAD, y_test_SAD = train_test_split(X, y_SAD, test_size=0.2, random_state=0)
```

```python
pipeline_SAD.fit(X_train_SAD, y_train_SAD)
```

```python
y_pred_SAD = pipeline_SAD.predict(X_test_SAD)
```

```python
print(f"Accuracy for SAD: {accuracy_score(y_test_SAD, y_pred_SAD) * 100:.2f}%")
print(f"Precision for SAD: {precision_score(y_test_SAD, y_pred_SAD, average='macro') * 100:.2f}%")
print(f"Recall for SAD: {recall_score(y_test_SAD, y_pred_SAD, average='macro') * 100:.2f}%")
print(f"F1 Score for SAD: {f1_score(y_test_SAD, y_pred_SAD, average='macro') * 100:.2f}%")
```

```python
pipeline_GAD = Pipeline(steps=[('preprocessor', preprocessor), ('model', RandomForestClassifier(n_estimators=917, random_state=0))])
```

```python
X_train_GAD, X_test_GAD, y_train_GAD, y_test_GAD = train_test_split(X, y_GAD, test_size=0.2, random_state=0)
```

```python
pipeline_GAD.fit(X_train_GAD, y_train_GAD)
```

```python
y_pred_GAD = pipeline_GAD.predict(X_test_GAD)
```

```python
print(f"Accuracy for GAD: {accuracy_score(y_test_GAD, y_pred_GAD) * 100:.2f}%")
print(f"Precision for GAD: {precision_score(y_test_GAD, y_pred_GAD, average='macro') * 100:.2f}%")
print(f"Recall for GAD: {recall_score(y_test_GAD, y_pred_GAD, average='macro') * 100:.2f}%")
print(f"F1 Score for GAD: {f1_score(y_test_GAD, y_pred_GAD, average='macro') * 100:.2f}%")
```

```python
def plot_confusion_matrix(confusion_matrix, title):
    plt.figure(figsize=(6, 4))
    sns.heatmap(confusion_matrix, annot=True, cmap='Blues', fmt='g')
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title(title)
    plt.show()
    
from sklearn.metrics import confusion_matrix

confusion_matrix_SAD = confusion_matrix(y_test_SAD, y_pred_SAD)
FP_SAD = confusion_matrix_SAD[0][1]
FN_SAD = confusion_matrix_SAD[1][0]
print('False Positives for SAD:', FP_SAD)
print('False Negatives for SAD:', FN_SAD)
plot_confusion_matrix(confusion_matrix_SAD, 'Confusion Matrix - SAD')
```

```python
confusion_matrix_GAD = confusion_matrix(y_test_GAD, y_pred_GAD)
FP_GAD = confusion_matrix_GAD[0][1]
FN_GAD = confusion_matrix_GAD[1][0]
print('False Positives for GAD:', FP_GAD)
print('False Negatives for GAD:', FN_GAD)
plot_confusion_matrix(confusion_matrix_GAD, 'Confusion Matrix - GAD')
```
